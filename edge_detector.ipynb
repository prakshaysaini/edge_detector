```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("container.png")
```

```python
while(img.shape[0]>1025):
    img=cv2.resize(img, (0, 0), fx=0.25, fy=0.25, interpolation=cv2.INTER_AREA)
img2=img.copy()
img3=img.copy()
img.shape
```

```python
vertical_convu_mat=np.array([[0.25,0.5,0.25],[0.0,0.0,0.0],[-0.25,-0.5,-0.25]])
vertical_convu_mat
```

```python
horizontal_convu_mat=np.array([[0.25,0,-0.25],[0.5,0,-0.5],[0.25,0,-0.25]])
horizontal_convu_mat
```

```python
def get_vertical_edges(i,j,img):
    sum=0.0
    for ii in range(-1,2):
        for jj in range(-1,2):
            if(i+ii>=0 and j+jj>=0 and i+ii<img.shape[0] and j+jj<img.shape[1]):
                sum += img[i+ii,j+jj]*vertical_convu_mat[ii+1][jj+1]
    return sum
```

```python
def get_horizontal_edge(i,j,img):
    sum=0.0
    for ii in range(-1,2):
        for jj in range(-1,2):
            if(i+ii>=0 and j+jj>=0 and i+ii<img.shape[0] and j+jj<img.shape[1]):
                sum += img[i+ii,j+jj]*horizontal_convu_mat[ii+1][jj+1]
    return sum
    
```

```python
for i in range(0,img.shape[0]):
    for j in range(0,img.shape[1]):
        img2[i,j]= get_vertical_edges(i,j,img)
        img3[i,j] = get_horizontal_edge(i,j,img)
```

```python
plt.title("Original image")
plt.imshow(img)
```

```python
plt.title("Vertical edges")
plt.imshow(img2)
```

```python
plt.title("Horizontal edges")
plt.imshow(img3)
```
